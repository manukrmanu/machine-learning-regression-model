```python
import numpy as np
import pandas as pd
df=pd.read_excel('/content/poly.xlsx')
df
```

```python
x=df.iloc[:,1:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
y_pred=model.predict(x)
y_pred
```

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score
print('mean error= ',mean_absolute_percentage_error(y,y_pred))
print('r2_score= ',r2_score(y,y_pred))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.plot(x,y_pred,color='red')
```

```python
from sklearn.preprocessing import PolynomialFeatures
x_poly=PolynomialFeatures(degree=3)
x_feature=x_poly.fit_transform(x)
x_feature
```

```python
model1=LinearRegression()
model1.fit(x_feature,y)
y_pred_poly=model1.predict(x_feature)
y_pred_poly
```

```python
plt.scatter(x,y,color='green')
plt.plot(x,y_pred_poly,color='red')
plt.xlabel('temperature')
plt.ylabel('PRESSURE')
plt.show()
```

```python
print('error is=',mean_absolute_percentage_error(y,y_pred_poly))
print("r2_score= ",r2_score(y,y_pred_poly))
```
