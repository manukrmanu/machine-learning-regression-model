```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/SalesPredictData (1).csv')
df
```

```python
df.isna().sum()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
y.ndim
```

```python
# importing seaborn
import seaborn as sns
sns.regplot(x=df['TV'],y=y)

```

```python
sns.regplot(x=df['Newspaper'],y=y)
```

```python
sns.regplot(x=df['Radio'],y=y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
print(model.predict([[150,55,72]]))
```

```python
lst=['TV','radio','newspaper']
j=0
for i in model.coef_:
  print('slope of',lst[j],'=',i)
  j+=1
```

```python
print("slope is ")
list(zip(x,model.coef_))
```

```python
print("constant is ",model.intercept_)
```

```python
df1=pd.DataFrame({'Actual_value':y_test,'predicted_value':y_pred,'difference':y_pred-y_test})
df1
```

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score,mean_squared_error
print('percentage error= ',mean_absolute_percentage_error(y_test,y_pred))
```

```python
print('r2_score= ',r2_score(y_test,y_pred))
```

```python
sq=mean_squared_error(y_test,y_pred)
print('root mean squared error= ',np.sqrt(sq))
```
