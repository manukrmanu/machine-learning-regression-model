```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/Salary_Data.csv')
df
```

```python
df.columns
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
x1=df['YearsExperience']
y1=df['Salary']
plt.scatter(x1,y1)
plt.grid()
plt.xlabel('experience')
plt.ylabel('salary')
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
x
```

```python
x1.ndim
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
x_train
```

```python
#model creation
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_test,y_pred,color='r')
plt.grid()
plt.xlabel('experience')
plt.ylabel('salary')
plt.show()
```

```python
print(model.predict([[8.5]]))
```

```python
#slope
print("slope is ",model.coef_)
```

```python
#intercept
print("constant is ",model.intercept_)
```

```python
dfp=pd.DataFrame({'actual_value':y_test,'predicted_value':y_pred,'diff':y_pred-y_test})
dfp
```

```python
# mean absolute error
from sklearn.metrics import mean_absolute_error
print('error is ',mean_absolute_error(y_test,y_pred))
```

```python
# mean absolute percentage error
from sklearn.metrics import mean_absolute_percentage_error
print('percentage error is ',mean_absolute_percentage_error(y_test,y_pred))
```

```python
# mean squared error
from sklearn.metrics import mean_squared_error
print('squared error is ',mean_squared_error(y_test,y_pred))
```

```python
#mean square error
from math import sqrt
from sklearn.metrics import mean_squared_error
x=mean_squared_error(y_test,y_pred)
print(sqrt(x))
```

```python
# r2 score
from sklearn.metrics import r2_score
print('r2 score error= ',r2_score(y_test,y_pred))
```
