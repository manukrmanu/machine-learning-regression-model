```python
import numpy as np
import pandas as pd
import seaborn as sns
df=pd.read_csv('/content/Student_Marks.csv')
df
```

```python
df.isna().sum()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
y.ndim
```

```python
sns.regplot(x=df['number_courses'],y=y)
```

```python
sns.regplot(x=df['time_study'],y=y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
print(model.predict([[7,5.6]]))
```

```python
print('slope is')
list(zip(x,model.coef_))
```

```python
print('constant is ',model.intercept_)
```

```python
df1=pd.DataFrame({'actual_value':y_test,'predicted_value':y_pred,'difference':y_pred-y_test})
df1
```

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score
print('mean percentage error= ',mean_absolute_percentage_error(y_test,y_pred))
print('r2_score= ',r2_score(y_test,y_pred))
```
